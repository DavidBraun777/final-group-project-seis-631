#  Are the sale prices and inventory listing going up or down overall in the US? 10 points 

```python
import pandas as pd
import numpy as np
import seaborn as sns                       #visualisation
import matplotlib.pyplot as plt             #visualisation
%matplotlib inline     
sns.set(color_codes=True)
```

```python
df = pd.read_csv("./for_sale_new_listings_month.csv")
df.head(20)
```

```python
df.tail(20)
```

```python
df.dtypes
```

```python
# I only care about the last years worth of data for now"
# Removing all columns before 10-31-2023."
start_date = '2018-01-31'
end_date = '2023-09-30'

# removing the first row also as it is just a sum of all the states
df = df.drop(index=0).reset_index(drop=True)

date_columns = [col for col in df.columns if pd.to_datetime(col, errors='coerce', format='%Y-%m-%d', exact=True) is not pd.NaT]

columns_to_drop = [col for col in date_columns if start_date <= col <= end_date]
df = df.drop(columns=columns_to_drop)

# I don't think we need the state in 2 columns so we'll remove the state abbreviation in the RegionName column
df["RegionName"] = df["RegionName"].apply(lambda x: x.split(',')[0])

# df.columns = df.columns.astype(str)
df.head(10)
```

```python
df.count()
```

```python
df = df.rename(columns={"SizeRank": "Rank", "RegionName": "City", "StateName": "State"})
df.head(10)
```

```python
# Let's start seeing if there are any duplicates we need to remove
df.shape
```

```python
duplicate_rows_df = df[df.duplicated()]
print("number of duplicate rows: ", duplicate_rows_df.shape)
```

```python
df.count()
```

```python
df = df.drop_duplicates()
df.count()
```

```python
# Now let's drop null or Nan values.
print(df.isnull().sum())
```

```python
df = df.dropna()
df.count()
```

```python
print(df.isnull().sum())
```
